```python
!pip install keras_ocr
import matplotlib.pyplot as plt
import keras_ocr
```

```python
pipline = keras_ocr.pipeline.Pipeline()
```

```python

images = [
    keras_ocr.tools.read(images) for images in [
        'https://docs.unity3d.com/Packages/com.unity.textmeshpro@3.2/manual/images/TMP_RichTextLineIndent.png',
        'https://i.stack.imgur.com/0Jl54.png'
    ]
]
```

```python
predictions=pipline.recognize(images)
```

```python
predictions
```

```python
detector = keras_ocr.detection.Detector()
recognizer = keras_ocr.recognition.Recognizer()
pipeline = keras_ocr.pipeline.Pipeline(detector=detector, recognizer= recognizer)
predictions = pipeline.recognize(images)
```

```python
fig,axs= plt.subplots(nrows=len(images), figsize=(20,20))
for ax, image, predictions in zip(axs, images, predictions):
  keras_ocr.tools.drawAnnotations(image=image, predictions=predictions, ax=ax)
```

```python
output=''
for text, box in predictions:
    output+= ' '+text
```
